### Énoncé

Vous étudiez le mouvement d'un projectile. Ce problème mobilise plusieurs modules de SciPy.

   t_mesure = np.array([0, 0.5, 1, 1.5, 2, 2.5, 3])

**Partie B - Résolution d'équation :**

## Exercice 2.1 : Analyse statistique d'un dataset médical
**Niveau : ⭐⭐ Intermédiaire**

### Énoncé
Vous analysez les données de tension artérielle de 30 patients avant et après un traitement.

**Données :**
```python
avant = [145, 138, 152, 148, 155, 142, 149, 150, 147, 151,
         143, 146, 154, 141, 148, 153, 144, 147, 150, 146,
         149, 152, 145, 148, 151, 143, 147, 150, 144, 149]

apres = [138, 132, 145, 140, 148, 135, 142, 143, 140, 144,
         136, 139, 147, 134, 141, 146, 137, 140, 143, 139,
         142, 145, 138, 141, 144, 136, 140, 143, 137, 142]
```

**Tâches :**

1. Calculez pour chaque groupe (avant/après) :
   - Moyenne, médiane, écart-type
   - Utiliser `scipy.stats.describe()`

2. Testez si la différence est statistiquement significative :
   - Test t de Student pour échantillons appariés (`ttest_rel`)
   - Interprétez la p-value (seuil α = 0.05)

3. Vérifiez la normalité des distributions :
   - Utilisez `shapiro` test
   - Tracez les histogrammes

4. Calculez la corrélation entre avant et après (`pearsonr`)

5. Rédigez une conclusion : le traitement est-il efficace ?


````python
from scipy import stats
avant=np.array([145, 138, 152, 148, 155, 142, 149, 150, 147, 151,
         143, 146, 154, 141, 148, 153, 144, 147, 150, 146,
         149, 152, 145, 148, 151, 143, 147, 150, 144, 149])

apres=np.array([138, 132, 145, 140, 148, 135, 142, 143, 140, 144,
         136, 139, 147, 134, 141, 146, 137, 140, 143, 139,
         142, 145, 138, 141, 144, 136, 140, 143, 137, 142])
print("Stats avant")
print(f"moyenne:{np.mean(avant)}")
print(f"médiane:{np.median(avant)}")
print(f"ecart-type:{np.std(avant)}")
print("Stats apres")
print(f"moyenne:{np.mean(apres)}")
print(f"médiane:{np.median(apres)}")
print(f"ecart-type:{np.std(apres)}")

t_statistiques,p_value=stats.ttest_rel(avant,apres)
print("test T")
print(f"t_statistiques={t_statistiques}")
print(f"p_value={p_value}")
if p_value<0.05:
  print("les données sont statistiquement significatives")
else:
  print("les données ne sont pas statistiquement significatives")

shapiro_avant=stats.shapiro(avant)
shapiro_apres=stats.shapiro(apres)
print("Test normalité")
print(f"avant:p={shapiro_avant.pvalue}{'(normal)'if shapiro_avant.pvalue>0.05 else '(non normal)'}")
print(f"apres:p={shapiro_apres.pvalue}{'(normal)'if shapiro_apres.pvalue>0.05 else '(non normal)'}")
corr,p_value=stats.pearsonr(avant,apres)
print("corrélation")
print(f"corr={corr}")
print(f"p_value={p_value}")

print("Conclusion")
reduction=np.mean(avant)-np.mean(apres)
print(f"reduction moyenne:{reduction}")
if p_value<0.05:
  print("le traitement est efficace")
else:
  print("le traitement n'est pas efficace")




````

## Exercice 2.2 : Analyse statistique avec scipy.stats
**Niveau : ⭐⭐ Intermédiaire**

### Énoncé

Vous analysez les résultats d'un examen passé par 200 étudiants.

**Tâches :**

1. **Génération des données :**
   ```python
   np.random.seed(42)
   notes = np.clip(np.random.normal(12, 3, 200), 0, 20)  # Notes entre 0 et 20
   ```

2. **Statistiques descriptives :**
   - Utilisez `stats.describe()` pour obtenir le résumé
   - Affichez la moyenne, la médiane et l'écart-type séparément

3. **Test de normalité :**
   - Utilisez `stats.normaltest()` pour vérifier si les notes suivent une loi normale
   - Interprétez le résultat (p-value)

4. **Ajustement d'une distribution :**
   - Utilisez `stats.norm.fit()` pour trouver les paramètres de la loi normale
   - Tracez l'histogramme des notes et superposez la PDF de la loi ajustée

5. **Calcul de probabilités :**
   - Quelle est la probabilité d'avoir une note > 15 ? (utilisez `cdf`)
   - Quelle note correspond au 90ème percentile ? (utilisez `ppf`)

````python
np.random.seed(42)
notes = np.clip(np.random.normal(12, 3, 200), 0, 20)

print("stats")
print(f"resumé : {stats.describe(notes)}")
print(f"moyenne : {np.mean(notes)}")
print(f"médiane : {np.median(notes)}")
print(f"écart-type : {np.std(notes)}")
#3
print("TEST NORMALITÉ")
k2 , p_value = stats.normaltest(notes)
print(f"p-value : {p_value}")
print(f"statistique : {k2}")
if p_value < 0.05:
    print("Les notes ne suivent pas une loi normale.")
else:
    print("Les notes suivent une loi normale.")
#4
mu , sigma = stats.norm.fit(notes)
print("AJUSTEMENT")
print(f"moyenne ajustée : {mu}")
print(f"ecart-type ajustée : {sigma}")
plt.figure(figsize=(10, 6))
plt.subplot(1,2,1)
plt.hist(notes, bins=20, color='blue', alpha=0.7)
x = np.linspace(0, 20,100)
pdf = stats.norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'red', linewidth=2)
plt.title('Histogramme des notes')
plt.xlabel('Notes')
plt.ylabel('Densité')
plt.legend()

plt.subplot(1,2,2)
plt.boxplot(notes , vert=True)
plt.title('Boxplot des notes')
plt.ylabel('Notes')
plt.show()
#5
print("PROBABILITES")
p_sup_15 =1- stats.norm.cdf(15, mu, sigma)
print(f"p(note > 15) = {p_sup_15}")
note_90 = stats.norm.ppf(0.9, mu, sigma)
print(f"note au 90ème percentile = {note_90}")
````

````python
import numpy as np
from scipy.sparse import csr_matrix , lil_matrix
from scipy.sparse.csgraph import connected_components , dijkstra , shortest_path
import sys
import networkx as nx

#1
print("matrice creuse")
n_etudiants = 1000
n_connexions = 7500
matrice = lil_matrix((n_etudiants, n_etudiants))
np.random.seed(42)
for _ in range(n_connexions):
    i , j = np.random.choice(n_etudiants, 2, replace=False)
    matrice[i, j] = 1
    matrice[j, i] = 1
matrice = csr_matrix(matrice)
print(f"Matrice créée : {n_etudiants} x {n_etudiants}")
print(f"Connexions : {matrice.nnz}")
#2
print("STATISTIQUES")
n_composantes , labels = connected_components(matrice)
print(f"Nombre de composantes connexes : {n_composantes}")
tailles = np.bincount(labels)
taille_max = np.max(tailles)
print(f"Taille de la plus grande composante : {taille_max}")
#3
print("Affichez les 10 étudiants")
degres = np.array(matrice.sum(axis=1)).flatten()
indices_tries = np.argsort(degres)[::-1]
for i in range(10):
    etudiant = indices_tries[i]
    print(f"Etudiant {etudiant} : {degres[etudiant]} ")
#4
print("plus court chemin")
etudiant1 , etudiant2 = np.random.choice(n_etudiants, 2, replace=False)
distances = shortest_path(matrice, indices=etudiant1)
distance= distances[etudiant2]
print(f"distance entre {etudiant1} et {etudiant2} : {distance}")
#5
print("comparaison mémoire")
matrice_dense = matrice.toarray()
mem_creuse = sys.getsizeof(matrice.data)+\
             sys.getsizeof(matrice.indices)+\
             sys.getsizeof(matrice.indptr)
mem_dense= matrice_dense.nbytes
print(f"mémoire creuse : {mem_creuse} octets")
print(f"mémoire dense : {mem_dense} octets")
print(f"gain : {mem_dense/mem_creuse}")

#--
print("visualisation")
sous_graphe = matrice[:50,:50].toarray()
G = nx.from_numpy_array(sous_graphe)
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=False, node_size=50, node_color='lightblue')
nx.draw_networkx_edges(G, pos, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=8, font_color='black')
plt.title('Sous-graphe du réseau social')
plt.show()
````

```python

# Bruit haute fréquence (ventilation, équipement)
bruit_hf = 0.8 * np.sin(2 * np.pi * 2000 * t)

# Bruit basse fréquence (ronronnement électrique 50 Hz)

# Bruit aléatoire

# Signal enregistré

```

**Tâches :**

1. Supprimez la tendance du signal avec `signal.detrend`

2. Appliquez la FFT pour analyser les fréquences :
   - Identifiez les pics de fréquence
   - Créez un spectre de puissance

3. Filtrez le signal :
   - Supprimez les fréquences < 80 Hz (bruit électrique)
   - Supprimez les fréquences > 400 Hz (bruit aigu)
   - Conservez la bande 80-400 Hz

4. Reconstruisez le signal avec la FFT inverse

5. Visualisez (4 sous-graphiques) :
   - Signal original bruité
   - Spectre avant filtrage
   - Spectre après filtrage
   - Signal nettoyé

6. Calculez le rapport signal/bruit (SNR) avant et après


````python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

#1
# Signal original : voix humaine (fréquences 100-300 Hz)
t = np.linspace(0, 10, 10000)
voix = np.sin(2 * np.pi * 150 * t) + 0.5 * np.sin(2 * np.pi * 250 * t)

# Bruit haute fréquence (ventilation, équipement)
bruit_hf = 0.8 * np.sin(2 * np.pi * 2000 * t)

# Bruit basse fréquence (ronronnement électrique 50 Hz)
bruit_bf = 0.6 * np.sin(2 * np.pi * 50 * t)

# Bruit aléatoire
bruit_random = np.random.normal(0, 0.3, len(t))

# Signal enregistré
signal_bruite = voix + bruit_hf + bruit_bf + bruit_random

#2
print("FFT")
n=len(t)
freqs = np.fft.fftfreq(n, d=t[1]-t[0])
fft_signal = np.fft.fft(signal_bruite)
puissance = np.abs(fft_signal)**2
pos_freq = freqs[:n//2]
pos_puissance = puissance[:n//2]
seuil=np.max(pos_puissance)*0.1
pics=pos_freq[pos_puissance>seuil]
valeurs_pics = pos_puissance[pos_puissance>seuil]
print("Pics de fréquence identifiés:")
for freq, val in zip(pics, valeurs_pics):
    print(f"  {freq:.0f} Hz - Puissance: {val:.2f}")
#3
print("filtrage")
fft_signal_filtre = fft_signal.copy()
fft_signal_filtre[np.abs(freq) <80] = 0
fft_signal_filtre[np.abs(freq) >400] = 0
signal_filtre = np.fft.ifft(fft_signal_filtre)
#4
print("reconstruction")
signal_filtre = np.fft.ifft(fft_signal_filtre).real
#5
fig, axs = plt.subplots(4, 1, figsize=(10, 15))
#5-1
axs[0].plot(t[:500], signal_bruite[:500], 'b')
axs[0].set_title('Signal bruité')
axs[0].set_xlabel('Temps (s)')
axs[0].set_ylabel('Amplitude')
#5-2
fft_avant = np.fft.fft(signal_bruite)
spectre_avant = np.abs(fft_avant)
axs[1].plot(freqs[:5000], spectre_avant[:5000], 'r')
axs[1].set_title('Spectre avant filtrage')
axs[1].set_xlabel('Fréquence (Hz)')
axs[1].set_ylabel('amplitude')
#5-3
fft_apres = np.fft.fft(signal_filtre)
spectre_apres = np.abs(fft_apres)
axs[2].plot(freqs[:5000], spectre_apres[:5000], 'g')
axs[2].set_title('Spectre après filtrage')
axs[2].set_xlabel('Fréquence (Hz)')
axs[2].set_ylabel('amplitude')
#5-4
axs[3].plot(t[:500], signal_filtre[:500], 'purple')
axs[3].set_title('Signal nettoyé')
axs[3].set_xlabel('Temps (s)')
axs[3].set_ylabel('Amplitude')

plt.tight_layout()
plt.show()
#6
print("SNR")
def snr(signal , bruit):
    puissance_signal = np.mean(signal**2)
    puissance_bruit = np.mean(bruit**2)
    return 10 * np.log10(puissance_signal / puissance_bruit)
bruit_avant = signal_bruite - voix
snr_avant = snr(voix , bruit_avant)
bruit_apres = signal_filtre - voix
snr_apres = snr(voix , bruit_apres)
print(f"SNR avant filtrage : {snr_avant:.2f} dB")
print(f"SNR après filtrage : {snr_apres:.2f} dB")
````

````python
from matplotlib import cm
from scipy import ndimage

#1
print("image synthétique")
image = np.ones((256, 256))*100
y, x= np.ogrid[:256, :256]
image[(x-100)**2 + (y-100)**2 <= 40**2] = 200
image[(x-150)**2 + (y-150)**2 <= 35**2] = 180
image[(x-80)**2 + (y-180)**2 <= 15**2] = 220
image += np.random.normal(0, 10, (256, 256))
#2
print("filtre gaussien")
sigma =2
image_filtre = ndimage.gaussian_filter(image, sigma)
#3
print("contours")
gradient = ndimage.sobel(image_filtre)
contours = gradient > 20
contours = ndimage.binary_closing(contours)
#4
print("segmentation")
labels, n_structures = ndimage.label(contours)
print(f"Nombre de structures détectées : {n_structures}")
tailles = ndimage.sum(contours, labels, range(1, n_structures+1))
print(f"aires : {tailles}")
#5
print("morphologie")
erode = ndimage.binary_erosion(contours)
dilate = ndimage.binary_dilation(contours)
#6
fig , axs = plt.subplots(2, 3 , figsize=(10, 8))
axs[0, 0].imshow(image , cmap='gray')
axs[0, 0].set_title('Image originale')
axs[0, 0].axis('off')
axs[0,1].imshow(image_filtre , cmap='gray')
axs[0,1].set_title('Image filtrée')
axs[0, 1].axis('off')
axs[0, 2].imshow(image , cmap='gray')
axs[0, 2].contour(contours , colors='red')
axs[0, 2].set_title('Contours détectés')
axs[0, 2].axis('off')
axs[1,0].imshow(labels , cmap='jet')
axs[1,0].set_title(f'Régions segmentées ({n_structures} structures)')
axs[1, 0].axis('off')
axs[1,1].imshow(erode , cmap='gray')
axs[1,1].set_title('Erosion')
axs[1, 1].axis('off')
axs[1,2].imshow(dilate , cmap='gray')
axs[1,2].set_title('Dilatation')
axs[1, 2].axis('off')

plt.tight_layout()
plt.show()
#--
print("centre de masse")
for i in range(1, n_structures+1):
  masque = labels == i
  centre_de_masse = ndimage.center_of_mass(image , masque)
  print(f"centre de masse de la structure {i} : {centre_de_masse}")
````

## Exercice 3.4 : Traitement d'image avec scipy.ndimage
**Niveau : ⭐⭐⭐ Avancé**

### Énoncé

Vous préparez des images binaires pour un algorithme de détection d'objets.

**Taches :**

1. **Création d'une image de test :**
   ```python
   np.random.seed(42)
   
   # Image 64x64 noire
   image = np.zeros((64, 64))
   
   # Ajout de formes géométriques (rectangles blancs)
   image[10:25, 10:30] = 1  # Rectangle 1
   image[35:55, 25:50] = 1  # Rectangle 2
   image[15:20, 45:55] = 1  # Rectangle 3
   
   # Ajout de bruit (pixels parasites)
   bruit = np.random.random((64, 64)) > 0.95  # 5% de pixels blancs aléatoires
   image_bruitee = np.maximum(image, bruit)
   ```

2. **Nettoyage morphologique :**
   - Appliquez `binary_opening()` pour supprimer les petits artefacts
   - Appliquez `binary_closing()` pour combler les petits trous

3. **Filtrage gaussien :**
   - Appliquez un filtre gaussien avec `gaussian_filter()` (sigma=1)
   - Binarisez le résultat avec un seuil de 0.5

4. **Visualisation (2x2 subplots) :**
   - Image originale
   - Image bruitée
   - Image après opening/closing
   - Image après filtrage gaussien

5. **Analyse :**
   - Comptez le nombre de pixels blancs dans chaque image
   - Commentez les différences entre les méthodes de nettoyage

````python
np.random.seed(42)

# Image 64x64 noire
image = np.zeros((64, 64))

# Ajout de formes géométriques (rectangles blancs)
image[10:25, 10:30] = 1  # Rectangle 1
image[35:55, 25:50] = 1  # Rectangle 2
image[15:20, 45:55] = 1  # Rectangle 3

# Ajout de bruit (pixels parasites)
bruit = np.random.random((64, 64)) > 0.95  # 5% de pixels blancs aléatoires
image_bruitee = np.maximum(image, bruit)

#2
image_ouverture = ndimage.binary_opening(image_bruitee)
image_fermeture = ndimage.binary_closing(image_ouverture)
#3
image_gauss = ndimage.gaussian_filter(image_fermeture, sigma=1)
image_bin = image_gauss > 0.5
#4
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
images = [image, image_bruitee, image_fermeture, image_bin]
titles = ['Image originale', 'Image bruitée', 'Image après opening/closing', 'Image après filtrage gaussien']
for ax, img , title in zip(axs.flatten(), images, titles):
    ax.imshow(img, cmap='gray')
    ax.set_title(title)
    ax.axis('off')
plt.tight_layout()
plt.show()
#5
print("Pixels blancs dans chaque image:")
noms = ['Image originale', 'Image bruitée', 'Image après opening/closing', 'Image après filtrage gaussien']
for mon , img in zip(noms , images):
  print(f"{mon} : {np.sum(img)} pixels blancs")
````
